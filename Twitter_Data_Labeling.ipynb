### Get Follower:Following Ratio

The higher the ratio, the higher chance the user is legitimate poster. They have more people following them to read tweets.

```python
# def getFollowerToFollowingRatio(client,id):
#     user = client.get_user(id=id,user_fields='public_metrics')
#     metrics = getUserMetrics(client,id)
#     followers = pd.json_normalize(metrics.json()['data'])['public_metrics.followers_count'][0]
#     following = pd.json_normalize(metrics.json()['data'])['public_metrics.following_count'][0]
#     return str(round(followers/following,2))
```

```python
def getFollowerToFollowingRatio(followers, following):
  if following == 0:
    ratio = 0
  else:
    ratio = followers / following
  return ratio
```

```python
df['Follower To Following'] = np.vectorize(getFollowerToFollowingRatio)(df['public_metrics.followers_count'], df['public_metrics.followers_count'])
df
```

```python
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer

sid = SentimentIntensityAnalyzer()

import re
import pandas as pd
import nltk
nltk.download('words')
words = set(nltk.corpus.words.words())
```

```python
def cleaner(text):
    tweet = re.sub("@[A-Za-z0-9]+","",str(text)) #Remove @ sign
    tweet = re.sub(r"(?:\@|http?\://|https?\://|www)\S+", "", str(text)) #Remove http links
    tweet = " ".join(tweet.split())
    tweet = tweet.lower()
    tweet = tweet.replace("#", "").replace("_", " ") #Remove hashtag sign but keep the text
    tweet = " ".join(w for w in nltk.wordpunct_tokenize(str(text))
         if w.lower() in words or not w.isalpha())
    return text
```

```python
df['tweet_clean'] = df['text'].apply(cleaner)
```

```python
word_dict = {'manipulate':-1,'manipulative':-1,'southwestiscancelled':-1,'cancelsouthwest':-1,
            'planecrash':-1,'ihatesouthwest':-1,'southwestsucks':-1,'cancelled':-1,'cancel culture':0.4,'teamdelta':-1,'teamsouthwest':1,
            'ilovesouthwest':1,'liar':-1, 'fake':-1,'cancel':-1, 'southweststinks':-1, 'worst flight':-1, 'hate southwest':-1, 'dislike southwest':-1}

import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer

sid = SentimentIntensityAnalyzer()
sid.lexicon.update(word_dict)

list1 = []
for i in df['tweet_clean']:
    list1.append((sid.polarity_scores(str(i)))['compound'])
```

```python
df['sentiment'] = pd.Series(list1)

def sentiment_category(sentiment):
    label = ''
    if(sentiment>0):
        label = 'positive'
    elif(sentiment == 0):
        label = 'neutral'
    else:
        label = 'negative'
    return(label)

df['sentiment_category'] = df['sentiment'].apply(sentiment_category)
```
