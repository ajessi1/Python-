```python
from plotly.offline import iplot
import plotly.graph_objects as go
trace0 = go.Box(y=df["CB9_ENRL_VACANCIES_INF"],name="Enrollment Vacancies Infants")
trace1 = go.Box(y=df["CB9_ENRL_VACANCIES_1YR"],name="Enrollment Vacancies 1 Year")
trace2 = go.Box(y=df["CB9_ENRL_VACANCIES_2YR"],name="Enrollment Vacancies 2 Year")
trace3 = go.Box(y=df["CB9_ENRL_VACANCIES_3YR"],name="Enrollment Vacancies 3 Year")
trace4 = go.Box(y=df["CB9_ENRL_VACANCIES_4YR"],name="Enrollment Vacancies 4 Year")
trace5 = go.Box(y=df["CB9_ENRL_VACANCIES_5YR"],name="Enrollment Vacancies 5 Year")
trace6 = go.Box(y=df["CB9_ENRL_VACANCIES_SA"],name="Enrollment Vacancies School Age")
layout = go.Layout(
    title="Enrollment Vacancies Across Age Groups",
    xaxis=dict(
        title="Age Group"
    ),
    yaxis=dict(
        title="Number of Vacancies"
    )
)

fig2 = [trace0, trace1, trace2,trace3, trace4, trace5, trace6]
iplot({'data':fig2,'layout': layout})

```
